```python
import pandas as pd

# Read the original CSV file
data = pd.read_csv("filmtv_movies - ENG.csv", sep=",")

# Get the number of rows in the original DataFrame
num_rows_original = data.shape[0]

# Clean the DataFrame // drop all NaN rows 
df = data.dropna(subset=["actors", "title",])

# Get the number of rows in the cleaned DataFrame
num_rows_cleaned = df.shape[0]

# Filter out movies with only one actor
df = df[df['actors'].str.split(',').map(len) > 1]

# Keep only movies from the United States
df = df[df['country'] == 'United States']

# Retain only the title and actors columns
df = df.loc[:, ['title', 'actors']]

# Replace commas with semicolons in the actors column
df['actors'] = data['actors'].str.replace(',', ';')

# Save the cleaned DataFrame to a CSV file
df.to_csv('movie.csv', index=False)
df
```
